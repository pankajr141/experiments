```python
import os
import pandas as pd

dfA = None
for filename in os.listdir('.'):
  if not filename.endswith('.csv'):
    continue
  df = pd.read_csv(filename)
  closing_colname = 'close_{}'.format(df['Symbol'].tolist()[0])
  df[closing_colname] = df['Close']
  df = df[['Date', closing_colname]]
  if dfA is None:
    dfA = df
    continue

  dfA = pd.merge(dfA, df, on=['Date'], how='inner')

dfA
```

### Granger Causality

https://en.wikipedia.org/wiki/Granger_causality

Ability of a timeseries to forecast another timeseries. Its a Statistical causality test which provide a way to investigate causality in term of prediction. It look at power of prediction.

can I enhance the enhance the estimation of y by looking at the past values of x and y ?


---


Lets take 2 coins BTC (bitcoin) and LTC (litecoin) and see if litecoin has any statistical impact on prediction on future values of BTC 

We will use **Granger Causality** present in statsmodels package for this purpose.

However we cannot use granger causality directly with out data, there are condition to check before generating granger causality
1. series should be stationary

```python
from statsmodels.tsa.stattools import adfuller


def _is_series_stationary(df, col, differencing_order):
  result = adfuller(df[col])
  p_value = result[1]
  stationary = False
  if p_value <= 0.05:
    stationary = True

  print("column: {}, do: {}, Test Statistics:{}, p-value: {}, stationary: {}".format(col, differencing_order, result[0], p_value, stationary))
  return stationary

# columns = ['close_BTC', 'close_LTC']
for col in dfA.columns:
  if col in ['Date']:
    continue
  differencing_order = 0
  while not _is_series_stationary(dfA, col, differencing_order):
    differencing_order += 1

    # Exit if even 2nd order differential is not stationary
    if differencing_order > 2:
      print("Unable to make column: {} stationary".format(col))
      break
    dfA[col] = dfA[col].diff().fillna(0)
```

```python
from statsmodels.tsa.stattools import grangercausalitytests
result = grangercausalitytests(dfA[['close_BTC', 'close_LTC']], maxlag=3)  # upto lag 3
```

since p value of test is p=0.0156 which is much lower then 0.05 that means the result is statistically significant or LTC has impact on future values of BTC.

### What other feature can help predict future values of BTC ? 



```python
columns_stats = []
for column in dfA.columns.tolist():
  if column in  ['Date', 'close_BTC']:
    continue

  max_lag = 3    # search upto max_lags
  result = grangercausalitytests(dfA[['close_BTC', column]], maxlag=max_lag)
  print(result)

  min_f_score = 1
  lag_num_with_min_f_score = -1
  column_included_in_forecast = False

  for i in range(1, max_lag + 1):
    f_score = result[i][0]['ssr_ftest'][1]
    if f_score < min_f_score and f_score < 0.05:
      column_included_in_forecast = True

    if f_score < min_f_score:
      min_f_score = f_score
      lag_num_with_min_f_score = i

  columns_stats.append({'col_name': column, 'column_included_in_forecast': column_included_in_forecast, 
                        'f_score': min_f_score, 'lag_num_with_min_f_score': lag_num_with_min_f_score})
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
pd.DataFrame(columns_stats)
```

### Transfer Entropy

transfer entropy is a non-parametric statistic measuring the amount of directed transfer of information between 2 random processes. It quantifies the statistical coherence between system evolving in time.

the entropy of a random variable is the average level of "information", "surprise", or "uncertainty" inherent to the variable's possible outcomes.
Since every feature has some entropy, our aim is to minimise it.
https://en.wikipedia.org/wiki/Entropy_(information_theory)

Can I reduce the entropy/uncertainty of in future values of y by looking at the past values of x and y ?

### Difference transfer entropy vs granger causality

**transfer entropy is**

1. Autoregressive approach to causal inference
1. more predice
2. more visually interpretable
3. Model free

**granger causality is**

1. information theoretic approach to causal inference 
1. Easier to automate
2. Less computationaly expensive
3. Easier to interpret the meaning of values obtained

**both approaches can** 
1. fail to distinguish causal interaction from correlation that arise due to similar governing equation or correlation that are induced by presence of common external forcing 
2. when system consist of more then 2 interacting processes they can return fake causality, i.e fail to distingish between direct and indirect causal interaction.
